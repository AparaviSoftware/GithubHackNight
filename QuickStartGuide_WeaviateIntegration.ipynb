# Milvus Weaviate - Quick Start Guide

This short guide will give you assistance on setting up transform action to a weaviate docker instance or cloud server. 

Requirements

- [Weaviate Vector Database Docker](https://weaviate.io/developers/weaviate/installation/docker-compose)
- [Weaviate Vector Database Cloud](https://weaviate.io/developers/wcs/create-instance)





### Adding an Integration

First you have to select the "Policies" tab and click "Add Integration"




Next, select the integration of your choice. In this case weviate cloud or docker...



Enter the parameters for the **Cloud Service**

You will need:

- Server URL
- API Key

*Service & Collection Name are defined by you.



**Remember to always save all changes!**





### Building an Action

Navigate to the actions tab and click on "Create Action"



Build a Weaviate Transform Action (here with all data that is owned by one particular user)



Click on "Save Action"

And click on "run" to initiate the transform action.
