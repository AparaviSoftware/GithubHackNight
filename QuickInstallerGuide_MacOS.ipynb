# APARAVI Quick Installer Guide - MacOS



Your installer file (which is a disk image) looks something like this

![Installer file.png](<attachment:Installer file.png>)

Double-click on the installer disk image. This will open up the installer. Double-click the Aparavi installer to start the installation.

![Installer open.png](<attachment:Installer open.png>)

Click "Allow" to give permissions to the installer

![Installer allow.png](<attachment:Installer allow.png>)

A terminal window will open for "Welcome to the APARAVI Installer for macOS" that explains what it will install and what you need (which is the sudo password for you Mac). If you do not have sufficient admin rights on your Mac, you will not be able to continue this installation.

Click "Enter" to start the installation.

![Installer terminal screen 1.png](<attachment:Installer terminal screen 1.png>)

Enter a password for the local MySQL installation and enter it twice to continue. Note that if you already have MySQL installed, it will skip re-installing it (such as in the image provided in the next step).

![Installer terminal screen 2 - msql password.png](<attachment:Installer terminal screen 2 - msql password.png>)

Next you will likely be prompted to enter your sudo password, then hit "Enter" to continue.

![Installer terminal screen 3 - sudo password.png](<attachment:Installer terminal screen 3 - sudo password.png>)

The installer starts uncompressing and extracting files, testing the MySQL connection, and then starts the download of the Aparavi software. This can take several minutes, depending on the bandwidth of your internet connection.

![Installer terminal screen 4 - download engine.png](<attachment:Installer terminal screen 4 - download engine.png>)

If you see a pop-up requesting you to allow administrator access to your Mac for the Installer, click "Allow"

![Installer terminal screen 5 - Allow admin.png](<attachment:Installer terminal screen 5 - Allow admin.png>)

When the Installer is finished, you will see this in the terminal window. Click "Enter" to close it.

![Installer terminal screen 6 - complete.png](<attachment:Installer terminal screen 6 - complete.png>)

The installer should automatically open up a browser window with the URL "http://localhost/on-prem-login"
