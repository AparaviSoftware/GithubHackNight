# Milvus Integration - Quick Start Guide

This short guide will give you assistance on setting up transform action to a milvus docker instance. 

Requirements

- [Milvus Vector Database Docker](https://milvus.io/docs/install_standalone-docker-compose.md)




### Adding an Integration

First you have to select the "Policies" tab and click "Add Integration"




Next, select the integration of your choice. In this case Milvus...



Enter the parameters, give a meaningful name to your integration as well as to your collection and press "Validate".



### Building an Action

Navigate to the actions tab.



### Building the tranform pipeline

Drag and drop the "Transform" icon on to the "Primary" field. Then curate the data you want to load into milvus by simply selecting the tag as a filter query. 



### Saving Your Changes

Make sure to save your changes.



### Populating the Milvus Vector Data Base

Once saved, click the "run" button to initiate the vectorization pipeline



### Observe your vectors getting populated

(Optional) Take a look into the [Attu Milvus GUI](https://github.com/zilliztech/attu)
